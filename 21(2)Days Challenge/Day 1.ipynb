### Question -1 : Add few names, one name in reach row, in "name.txt"

a. count no of names
b. count all names starting with vowel
c. find longest name

```python
with open("names.txt", "r") as f:
    names = f.readlines()
    print("Total Names: ", len(names))
```

b) count all names starting with vowel

```python
count = 0 
vowels = "AEIOUaeiou"
with open("names.txt", "r") as f:
    for name in f:
        if name.strip()[0] in vowels:
            count += 1

print("Names starting with vowels: ", count)
```

c) find longest name

```python
with open("names.txt", "r") as f:
    names = [name.strip() for name in f]
    longest = max(names, key=len)

print("Longest name:", longest)
```

### Question 2: Store integers in a file

a) find the max number

```python
with open("integers.txt", "r") as f:
    nums = list(map(int, f.read().split()))

print("Maximum: ", max(nums))
```

b) find average of all numbers

```python
avg = sum(nums) / len(nums)
print("Average: ", avg)
```
