### 1) Write a Python program to input marks and credits of 5 subjects. Calculate the SGPA/CGPA using weighted average formula and display the final CGPA.

```python
total_points = 0
total_credits = 0

for i in range(2):
    marks = float(input("enter marks for subject{i+1}:"))
    credit = int(input("enter credit for subject{i+1}:"))

    grade_point = marks / 10
    total_points += grade_point * credit
    total_credits += credit

cgpa = total_points / total_credits
print("Final CGPA: ", round(cgpa, 2))
```

### 2) Write a Python program to input:

total classes held
classes attended

Display the attendance percentage.
Also check whether the student satisfies the 75% attendance criteria.
If not, display how many more classes must be attended continuously to reach 75%.

```python
total = int(input("enter total classes : "))
attended = int(input("enter attended classes : "))

attendance = (attended / total) * 100
print("attendance Percentage : ", round(attendance, 2), "%")

if attendance >= 75:
    print("Attendance criteria satisfied")
else:
    needed = 0
    while ((attended + needed) / (total + needed)) * 100 < 75:
        needed += 1
    print("Need to attend", needed, "more classes continuously")
```

### 3) Write a Python program to input the expenses of 4 hostel friends on food, travel, and miscellaneous items. Calculate the total expense and display how much each person should pay equally.

```python
expenses = []

for i in range(4):
    amount = float(input("enter expenses of friend {i+1}: "))
    expenses.append(amount)

total = sum(expenses)
split = total / 4

print("Total expenses : ", total)
print("each person should pay: ", split)
```

### 4) Write a Python program using classes and objects to manage a small library system.
The program should allow:

adding a book
issuing a book
returning a book
displaying all available books

```python
class library:
    def __init__(self):
        self.book = []

    def add_book(self, book):
        self.book.append(book)

    def issue_book(self, book):
        if book in self.book:
            self.book.remove(book)
        else:
            print("Book not available")

    def return_book(self, book):
        self.book.append(book)
        print("Book returned")

    def display_books(self):
        print("Available books : ", self.book)

lib = library()
lib.add_book("Python")
lib.add_book("DBMS")
lib.add_book("DSA")
lib.display_books()
lib.issue_book("Python")
lib.display_books()
lib.return_book("Python")
```
