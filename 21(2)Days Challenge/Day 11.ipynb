### 1) Write a Python program to input the exam date from the user and calculate the number of days remaining from today’s date.

```python
from datetime import datetime

exam_date = input("enter exam date (YYYY-MM-DD): ")
exam = datetime.strptime(exam_date, "%Y-%m-%d")

today = datetime.today()

days_left = (exam - today).days
print("Days left for examm: ", days_left)
```

### 2) Write a Python program to check whether a password is strong or weak.
A strong password must contain:

at least 8 characters
one uppercase letter
one lowercase letter
one digit
one special character

```python
password = input("enter password: ")

upper = lower = digit = special = False

for ch in password:
    if ch.isupper():
        upper = True
    elif ch.islower():
        lower = True
    elif ch.isdigit():
        digit = True
    else:
        special = True

if len(password) >= 8 and upper and lower and digit and special:
    print("Strong password")
else: 
    print("Weak Password")
```

### 3) Write a Python program to create a canteen billing system. Display a menu with item prices, accept quantity from the user, and generate the final bill including 5% GST.

```python
menu = {"Burger" : 50, "Pizza" : 100, "Tea" : 20}

total = 0
for item, price in menu.items():
    qty = int(input("enter quantity for {item}:"))
    total += price * qty

gst = total * 0.05
final_bill = total + gst

print("Total Bill: ", final_bill)
```

### 4) Write a Python program to create a multiple-choice quiz application for 5 questions.Display the final score at the end.

```python
questions = {
    "Python is developed by?" : "Guido",
    "Keywords for function?":"def",
    "List uses which brackets?" : "[]",
    "Loop for fixed iterations?" : "for",
    "Tuple is immutable ? ": "yes"
}

score = 0
for q, ans in questions.items():
    user = input(q)
    if user.lower() == ans.lower():
        score += 1

print("Final score : ", score)
```
