### Write a Python program to input marks of 10 students in a subject.
### Display:

highest marks
lowest marks
class average
number of failed students (marks below 40)

```python
marks = []

for i in range(10):
    m = int(input(f"Enter marks of student {i+1}: "))
    marks.append(m)

print("Highest:", max(marks))
print("Lowest:", min(marks))
print("Average:", sum(marks) / len(marks))

fail = 0
for m in marks:
    if m < 40:
        fail += 1

print("Failed Students:", fail)
```

### Write a Python program to check whether a student is eligible for placement.
### Eligibility conditions:

CGPA ≥ 7
no active backlogs
at least 1 internship

Display whether the student is eligible or not.

```python
cgpa = float(input("enter CGPA:"))
backlogs = int(input("enter active backlogs"))
intership = input("Do you have internship? (y/n):")

if cgpa >= 7 and backlogs == 0 and intership.lower() == "yes":
    print("Eligible for placement")
else:
    print("Not eligible")
```

### Write a Python program using classes to create a to-do list.
Each task should store:

task name
deadline
priority

Display all pending tasks sorted by priority.

```python
tasks = []

for i in range(3):
    name = input("enter task:")
    deadline = input("enter deadline: ")
    priority = int(input("enter priority:"))
    tasks.append((priority, name, deadline))

tasks.sort()

for task in tasks:
    print(task)
```

### Write a Python program to convert Indian Rupees into US Dollars and Euros using fixed conversion rates.

```python
inr = float(input("Enter INR: "))

usd = inr / 83
eur = inr / 90

print("USD :", round(usd, 2))
print("EUR : ", round(eur, 2))
```
