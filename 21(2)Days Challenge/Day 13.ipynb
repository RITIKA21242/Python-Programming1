### Write a Python program to create a mini train reservation system. Store passenger details and allocate seats based on availability.

```python
seats = 5

while seats > 0:
    name = input("enter passenger name :")
    print("Seat booked for : ", name)
    seats -= 1
    print("Remaining seats:", seats)
```

### Write a Python program to take votes for 3 candidates and display the winner based on maximum votes.

```python
votes = {"A":0, "B": 0, "C": 0}

for i in range(5):
    vote = input("vote for A/ B/ C: ").upper()
    if vote in votes:
        votes[vote] += 1

winner = max(votes, key = votes.get)
print("Winner is: ", winner)
```

### Write a Python program to create a file-based student record system.
### The program should allow the user to:

add a student record
display all student records
search a student by roll number
update marks of a student

### Store the data in a text file.

```python
def add_record():
    with open("students.txt", "a") as file:
        roll = input("Enter Roll No: ")
        name = input("Enter Name: ")
        marks = input("Enter Marks: ")
        file.write(f"{roll},{name},{marks}\n")

def display_records():
    with open("students.txt", "r") as file:
        print(file.read())

def search_record():
    roll_search = input("Enter Roll No to search: ")
    with open("students.txt", "r") as file:
        for line in file:
            roll, name, marks = line.strip().split(",")
            if roll == roll_search:
                print("Record Found:", roll, name, marks)

def update_marks():
    roll_update = input("Enter Roll No to update: ")
    lines = []

    with open("students.txt", "r") as file:
        lines = file.readlines()

    with open("students.txt", "w") as file:
        for line in lines:
            roll, name, marks = line.strip().split(",")
            if roll == roll_update:
                marks = input("Enter New Marks: ")
            file.write(f"{roll},{name},{marks}\n")

# Example menu
while True:
    print("\n1.Add 2.Display 3.Search 4.Update 5.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        add_record()
    elif choice == "2":
        display_records()
    elif choice == "3":
        search_record()
    elif choice == "4":
        update_marks()
    elif choice == "5":
        break
    else:
        print("Option not listed in the choices")
```
