### Write a Python program to manage product stock in a store.
The system should:

store item names with quantities
reduce stock when items are sold
automatically identify items below restock threshold
display items that need urgent refill

```python
stock = {"pen" : 12, "notebook": 4, "marker": 2}

threshold = 5

sold_item = input("enter sold item : ").lower()
qty = int(input("enter quantity sold : "))

if sold_item in stock :
    stock[sold_item] -= qty

print("items needing restock:")
for item, qty in stock.items():
    if qty < threshold:
        print(item, "->", qty)
```

### Write a Python program to read a text log file and count the frequency of:

ERROR
WARNING
INFO

Display the most frequent log type.

```python
log_data = ["error", "info", "warning", "error", "error", "info"]

count = {"error":0, "warning":0, "info":0}

for log in log_data:
    count[log] += 1

print("Log Frequency:", count)
print("Most Frequent:", max(count, key=count.get))
```

### Write a Python program to store attendance of students for multiple days and generate:

total attendance per student
class attendance percentage
most irregular student

```python
attendance = {"Riya" : [1, 1, 0, 1, 0],
              "Aman" : [1, 0, 0, 1, 1]
              }

for student, days in attendance.items():
    total = sum(days)
    percent = (total / len(days)) * 100
    print(student, "Attendance:", percent, "%")
```

### Write a Python program to allocate classroom seats for students during exams.
The program should:

take roll numbers
assign bench numbers
avoid duplicate seat assignment
display final seating chart


```python
students = ["101", "102", "103", "104"]
seating = {}

bench = 1
for roll in students:
    seating[roll] = bench
    bench += 1

print("Seating Chart:")
for roll, seat in seating.items():
    print("Roll: ", roll, "Bench:", seat)
```
