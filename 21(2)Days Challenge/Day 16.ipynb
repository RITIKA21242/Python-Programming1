### 1) Hostel Complaint Portal

Write a Python program using classes and objects to design a hostel complaint portal that stores complaint details such as room number, complaint type, urgency level, and current status.

```python
class complaint:
    def __init__ (self, room_no, complaint_type, urgency, status):
        self.room_no = room_no
        self.complaint_type = complaint_type
        self.urgency = urgency
        self.status = status

    def display(self):
        print("Room: ", self.room_no)
        print("Type :", self.complaint_type)
        print("Urgency :", self.urgency)
        print("Status :", self.status)

c1 = complaint(204, "Water Leakage", "High", "Pending")
c1.display()
```

### 2) Event Registration Manager

Write a Python program using classes and objects to implement an event registration management system that stores participant details, department, event name, and attendance information.

```python
class EventParticipant:
    def __init__(self, part_id, event_nme, depart, attendance):
        self.part_id = part_id
        self.event_nme = event_nme
        self.depart = depart
        self.attendance = attendance

    def display(self):
        print("ID:", self.part_id)
        print("Event:", self.event_nme)
        print("Department :", self.depart)
        print("Attendance :", self.attendance)

p1 = EventParticipant(101, "Hackathon", "CSE", "Present")
p1.display()
```

### 3) Lab Equipment Borrow System

Write a Python program using classes and objects to create a lab equipment borrowing system that manages item details, borrower information, and return status.

```python
class Equipment:
    def __init__(self, item_id, item_name, borrower, status):
        self.item_id = item_id
        self.item_name = item_name
        self.borrower = borrower
        self.status = status

    def display(self):
        print("Item ID: ", self.item_id)
        print("Item:", self.item_id)
        print("Borrowed by ", self.borrower)
        print("Status: ", self.status)

e1 = Equipment(501, "Oscilloscope", "Ritika", "Issued")
e1.display()
```

### 4) Wi-Fi Usage Session Logger

Write a Python program using classes and objects to simulate a Wi-Fi session logging system that stores user login details, session timings, and total data usage.

```python
class session:
    def __init__(self, usernme, login_time, logout_time, data_used):
        self.usernme = usernme
        self.login_time = login_time
        self.logout_time = logout_time
        self.data_used = data_used

    def display(self):
        print("User :", self.usernme)
        print("Login", self.login_time)
        print("Logout :", self.logout_time)
        print("Data Used :", self.data_used, "GB")

s1 = session("Ritika", "10:00 AM", "12:00 PM", 2.5)
s1.display()
```
