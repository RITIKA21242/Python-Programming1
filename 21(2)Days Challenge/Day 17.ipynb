### 5) Portfolio Project Tracker

Write a Python program using classes and objects to design a portfolio project tracking system that stores project title, technology stack, completion percentage, and repository link.

```python
class project:
    def __init__(self, title, tech_stack, completion, repo):
        self.title = title
        self.tech_stack = tech_stack
        self.completion = completion
        self.repo = repo

    def display(self):
        print("Title : ", self.title)
        print("Tech Stack :", self.tech_stack)
        print("Completion :", self.completion, "%")
        print("Repository : ", self.repo)

p1= project(
    "College ERP",
    "Python, MySQL",
    80,
    "github.com/erp-project"
)
p1.display()
```

### v6) Smart Notes Version Manager

Write a Python program using classes and objects to implement a notes version management system that stores note title, content, version details, and modification date.

```python
class Note:
    def __init__(self, title, content, version, modified_date):
        self.title = title
        self.content = content
        self.version = version
        self.modified_date = modified_date

    def display(self):
        print("Title:", self.title)
        print("Content:", self.content)
        print("Version:", self.version)
        print("Modified:", self.modified_date)

n1 = Note(
    "DBMS Notes",
    "Functional Dependency",
    3,
    "13-Apr-2026"
)
n1.display()
```

### 7) Gym Membership System

Write a Python program using classes and objects to create a gym membership management system that stores member details, plan type, and renewal status.

```python
class member:
    def __init__(self, member_id, name, plan, renewel_status):
        self.member_id = member_id
        self.name= name
        self.plan = plan
        self.renewel_status = renewel_status

    def display(self):
        print("ID:", self.member_id)
        print("Name :", self.name)
        print("Plan :", self.plan)
        print("Renewal :", self.renewel_status)

m1= member(301, "Ritika", "Premium", "Active")
m1.display()
```

### 8) Bus Pass Management System

Write a Python program using classes and objects to design a bus pass management system for students that stores route number, validity date, and active status.

```python
class bussPass:
    def __init__(self, stud_id, route_no, validity, status):
        self.stud_id = stud_id
        self.route_no = route_no
        self.validity = validity
        self.status = status

    def display(self):
        print("Student ID", self.stud_id)
        print("Route :", self.route_no)
        print("Validity :", self.validity)
        print("Status :", self.status)

b1 = bussPass(101, "Route_5", "30-April-2026", "Active")
b1.display()
```
