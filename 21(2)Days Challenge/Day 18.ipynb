### 9) Classroom Asset Monitoring System

Write a Python program using classes and objects to manage classroom assets such as projectors, systems, and smart boards, including their availability and maintenance status.

```python
class asset:
    def __init__(self, asset_id, asset_nme, availability, maintenance):
        self.asset_id = asset_id
        self.asset_nme = asset_nme
        self.availability = availability
        self.maintenance = maintenance

    def display(self):
        print("Asset ID:", self.asset_id)
        print("Asset_nme :", self.asset_nme)
        print("Availability : ", self.availability)
        print("Maintenance : ", self.maintenance)

a1 = asset(901, "Projector", "Available", "Good")
a1.display()
```

### 10) Digital Certificate Issuance System

Write a Python program using classes and objects to implement a digital certificate management system that stores student details, event name, certificate ID, and issue status.

```python
class Certificate:
    def __init__(self, student_name, event_name, certificate_id, issue_status):
        self.student_name = student_name
        self.event_name = event_name
        self.certificate_id = certificate_id
        self.issue_status = issue_status

    def display(self):
        print("Student:", self.student_name)
        print("Event:", self.event_name)
        print("Certificate ID:", self.certificate_id)
        print("Status:", self.issue_status)

c1 = Certificate("Riya", "Hackathon", "CERT101", "Issued")
c1.display()
```

### 11) Laptop Service Center Management System

Write a Python program using classes and objects to design a laptop service center management system that stores customer details, device model, issue type, service status, and estimated delivery date.

```python
class LaptopService:
    def __init__(self, customer, model, issue, status, delivery):
        self.customer = customer
        self.model = model
        self.issue = issue
        self.status = status
        self.delivery = delivery

    def display(self):
        print("Customer : ", self.customer)
        print("Model : ", self.model)
        print("Issue ", self.issue)
        print("Status :", self.status)
        print("Delivery : ", self.delivery)

l1 = LaptopService(
    "Ritika",
    "Lenevo IdeaPad",
    "Battery Issue",
    "In Progress",
    "20-April-2026"
)
l1.display()
```

### 12) Online Examination Attempt Tracker

Write a Python program using classes and objects to implement an online examination attempt tracking system that stores student details, subject name, attempt count, score, and completion status.

```python
class ExamAttempt:
    def __init__(self, student, subject, attempts, score, status):
        self.student = student
        self.subject = subject
        self.attempts = attempts
        self.score = score
        self.status = status

    def display(self):
        print("Student :", self.student)
        print("Subject :", self.subject)
        print("Attempts : ", self.attempts)
        print("Score :", self.score)
        print("Status :", self.status)

e1 = ExamAttempt("Ritika", "Python", 2, 88, "Completed")
e1.display()
```
