### 1. Write a Python program using classes and objects to design a shopping cart that allows multiple products to be added, removed, and calculates the final payable amount dynamically.

```python
class shoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, name, price):
        self.items.append((name, price))

    def remove_item(self, name):
        self.items = [item for item in self.items if item[0] != name]

    def total_bill(self):
        total = sum(price for _, price in self.items)
        print("Final Payable Amount :", total)

cart = shoppingCart()
cart.add_item("Mouse", 500)
cart.add_item("Keyboard", 1200)
cart.remove_item("Mouse")
cart.total_bill()
```

### 2. Write a Python program using classes and objects to implement a hostel room booking system where room availability changes after booking and checkout operations.

```python
class hostelRoom:
    def __init__(self, total_rooms):
        self.available_rooms = total_rooms

    def book_room(self):
        if self.available_rooms > 0:
            self.available_rooms -= 1
            print("Room Booked Successfully")
        else:
            print("No rooms Available")

    def checkout(self):
        self.available_rooms += 1
        print("Checkout completed")

    def show_status(self):
        print("Available Rooms :", self.available_rooms)

hostel = hostelRoom(5)
hostel.book_room()
hostel.book_room()
hostel.checkout()
hostel.show_status()
```

### 3. Write a Python program using classes and objects to create a movie ticket booking system in which seats are booked dynamically and seat availability is updated after each booking.

```python
class movieTicket:
    def __init__(self, total_seats):
        self.available_seats = total_seats

    def book_tickets(self, seats):
        if seats <= self.available_seats:
            self.available_seats -= seats
            print(seats, "seat(s) booked")
        else:
            print("Not enough seats available")

    def show_seats(self):
        print("Available Seats :", self.available_seats)


movie = movieTicket(20)
movie.book_tickets(3)
movie.book_tickets(5)
movie.book_tickets(4)
movie.show_seats()
```

### 4. Write a Python program using classes and objects to simulate a library system that calculates fine based on the number of delayed days after book return.

```python
class libraryBook:
    def __init__(self, title):
        self.title = title

    def calculate_fine(self, late_days):
        fine = late_days * 5
        print("Book:", self.title)
        print("Fine Amount:", fine)

book = libraryBook("DBMS")
book.calculate_fine(4)
```
