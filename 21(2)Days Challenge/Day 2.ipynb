### Question 3 : Assume a file city.txt with details of 5 cities in given  format (cityname popuation(in lakhs), area(in sq KM))

Example: 

Dehradun 5.78 308.20

Delhi 190 1484

### Open file city.txt and read to: 

a) Display details  of all cities


```python
with open("city.txt", "r") as f:
    for line in f:
        name, pop, area = line.split()
        print(f"Name : {name}, Population : {pop} lakhs, Area : {area} sq km")
```

b) Display city names with population more than 10 lakhs

```python
with open("city.txt", "r") as f:
    for line in f:
        name, pop, area = line.split()
        if float(pop) > 10:
            print(name)
```

c) Display sum of areas of all cities

```python
total_area = 0
with open("city.txt", "r") as f:
    for line in f:
        name, pop, area = line.split()
        total_area += float(area)

print("Total Area : ", total_area)
```

### Question 4 : Input two values from user where the first line contains N, the number of test  cases. the next N lines contain the space separated values of a and b. Perform integer division and print a/b. Handle exception in case of ZeroDivision or Value error.

### Sample Input: 

1. 0
2. $
3. 1

### Sample output:
 
Error code : integer division or modulo by zero
Error code: invalid literal for int() with base 10: '$' 3

```python
n = int(input())

for _ in range(n):
    try:
        a, b = input().split()
        result = int(a) // int(b)
        print(result)

    except ZeroDivisionError as e:
        print("Error Code:", e)

    except ValueError as e:
        print("Error Code:", e)
```
