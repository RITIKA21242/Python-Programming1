### 1. Write a Python program using classes and objects to manage a quiz system where the score updates dynamically after each correct answer.

```python
class quiz:
    def __init__(self):
        self.score = 0

    def correct_answer(self):
        self.score += 1

    def wrong_answer(self):
        pass

    def show_score(self):
        print("Final Score :", self.score)

quiz = quiz()
quiz.correct_answer()
quiz.correct_answer()
quiz.wrong_answer()
quiz.show_score()
```

### 2. Write a Python program using classes and objects to track workout sessions, calories burned, and total workout duration for a gym member.

```python
class workout:
    def __init__(self):
        self.total_minutes = 0
        self.calories = 0

    def add_session(self, minutes, calories):
        self.total_minutes += minutes
        self.calories += calories

    def show_summary(self):
        print("Total Workout Duration:", self.total_minutes, "minutes")
        print("Total Calories Burned:", self.calories)

gym = workout()
gym.add_session(30, 200)
gym.add_session(45, 350)
gym.show_summary()
```

### 3. Write a Python program using classes and objects to simulate battery percentage changes based on charging and usage operations.

```python
class laptopBattery:
    def __init__(self, battery = 100):
        self.battery = battery

    def use_laptop(self, percent):
        self.battery -= percent
        if self.battery < 0:
            self.battery = 0

    def charge_laptop(self, percent):
        self.battery += percent
        if self.battery > 100:
            self.battery = 100

    def show_battery(self):
        print("Battery Percentage:", self.battery, "%")


lap = laptopBattery()
lap.use_laptop(30)
lap.charge_laptop(10)
lap.show_battery()
```

### 4. Write a Python program using classes and objects to calculate cab fare dynamically based on distance travelled and waiting time.

```python
class cabFare:
    def __init__(self, base_fare):
        self.base_fare = base_fare

    def calculate_dare(self, distance, waiting_time):
        fare = self.base_fare + (distance * 15) + (waiting_time * 2)
        print("Total Fare :", fare)

cab = cabFare(50)
cab.calculate_dare(10, 5)
```
