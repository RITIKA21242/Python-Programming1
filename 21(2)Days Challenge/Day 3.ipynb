### Question 5 : Create multiple suitable exceptions for a file handling program

```python
try:
    with open("Datas.txt", "r") as f:
        content = f.read()
        print(content)

except FileNotFoundError:
    print("Error: File not Found")

except PermissionError:
    print("Error: Permission Denied")

except Exception as e:
    print ("Some other error occured: ", e)
```

### Question 6: Count uppercase, lowercase & digits in a file
Question:

Read a file text.txt and count:

Number of uppercase letters
Number of lowercase letters
Number of digits

```python
upper = lower = digit = 0

with open("text.txt", "r") as f:
    data = f.read()

for ch in data:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1
    elif ch.isdigit():
        digit += 1

print("Uppercase Count :", upper)
print("Lowercase Count: ", lower)
print("Digits count: ", digit)
```

### Question 7: Copy content from one file to another (skip blank lines)
Question:

Copy content from file1.txt to file2.txt but:
Do NOT copy empty lines

```python
with open("file1.txt", "r") as f1, open("file2.txt", "w") as f2:
    for line in f1:
        if line.strip() != "":
            f2.write(line)

print("File copied without Blank Lines")
```

### Question 8: Count number of words in each line
Question:

Read data.txt and print:
number of words in each line

```python
with open("data.txt", "r") as f:
    for line in f:
        words = line.split()
        print("words in line: ", len(words))
```
