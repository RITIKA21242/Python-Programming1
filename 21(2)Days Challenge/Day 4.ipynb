### 1) Find the largest and smallest element

### Given an array of n elements, find:

largest number
smallest number

### Example: [4, 8, 1, 9, 3] → max = 9, min = 1

```python
n = int(input("Enter size: "))
arr = list(map(int, input("Enter elements: ").split()))
print("Largest: ", max(arr))
print("Smallest : ", min(arr))
```

### 2) Find the second largest element

### Find the second highest number in an array without sorting.

### Example: [10, 20, 5, 8] → 10

```python
n = int(input("Enter size: "))
arr = list(map(int, input("enter elements: ").split()))

largest = second = -9999

for num in arr:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest :
        second = num

print("Second Largest = ", second)
```

### 3) Reverse an array

### Reverse the elements of the array.

### Example: [1,2,3,4,5] → [5,4,3,2,1]

```python
n = int(input("enter size: "))
arr = list(map(int, input("enter elements : ").split()))

print("Reversed array: ", arr[::-1])
```

### 4) Count even and odd numbers

### Count how many even and odd elements are present.

### Example: [1,2,3,4,5,6]

Even = 3

Odd = 3

```python
n = int(input("enter size: "))
arr = list(map(int, input("enter ellements: ").split()))

even = 0
odd = 0

for num in arr: 
    if num % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even : ", even)
print("Odd: ", odd)
```
