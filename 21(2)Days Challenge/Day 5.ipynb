### 5) Sum and average of array

### Find:

sum of all elements
average of elements

### Example: [2,4,6,8]
Sum = 20
Average = 5



```python
n = int(input("enter size: "))

arr = list(map(int, input("enter elements: ").split()))

total = sum(arr)
avg = total / n

print("Sum : ", total)
print("Average: ", avg)
```

### 6) Search an element

### Take an element from the user and check whether it exists in the array.

### If found, print its position too.



```python
n = int(input("enter size: "))

arr = list(map(int, input("enter elements : ").split()))

key = int(input("elements to search: "))

if key in arr:
    print("Found position: ", arr.index(key)+1)
else:
    print("Index Not Found")
```
