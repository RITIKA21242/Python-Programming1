### 1) Create a Student class

### Create a class Student with:

name
roll number
marks

### Create an object and display student details.

```python
class student:
    def __init__ (self, name, roll, marks):
        self.name = name
        self.roll = roll
        self.marks = marks
    
    def display(self):
        print("Name:", self.name)
        print("Roll No: ", self.roll)
        print("Matks : ", self. marks)

s1 = student("Aman", 101, 89)
s1.display()
```

### 2) Use Constructor

### Create a class Car with constructor to initialize:

brand
model
price

### Create object and print details.

```python
class car:
    def __init__ (self, brand, model, price):
        self.brand = brand
        self.model = model
        self.price = price
    
    def show(self):
        print("Brand: ", self.brand)
        print("Model : ", self.model)
        print("Price: ", self.price)

c1 = car("BMW", "X5", 8500000)
c1.show()
```

### 3) Create a Rectangle class

### Make a class with:

length
breadth
method to calculate area
method to calculate perimeter

```python
class rectangle:
    def __init__ (self, length, breadth):
        self.length = length
        self.breadth = breadth
    
    def area(self):
        return self.length * self.breadth
    
    def perimeter(self):
        return 2*(self.length + self.breadth)

r1 = rectangle(10, 5)
print("Area =", r1.area())
print("Perimeter = ", r1.perimeter())
```

### 4) Single Inheritance

### Create:

class Person
class Student(Person)

### Inherit name from Person and add marks in Student.

```python
class Person:
    def __init__(self, name):
        self.name = name

    def show_name(self):
        print("Name:", self.name)


class Student(Person):
    def __init__(self, name, marks):
        super().__init__(name)
        self.marks = marks

    def show_marks(self):
        print("Marks:", self.marks)


s1 = Student("Ritika", 95)
s1.show_name()
s1.show_marks()
```
