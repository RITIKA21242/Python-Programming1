### 5) Multilevel Inheritance

### Create:

Animal
Dog(Animal)
Puppy(Dog)

### Show inherited methods from all levels.

```python
class animal: 
    def sound(self):
        print("Animal makes sound")

class dog(animal):
    def bark(self):
        print("Dog barks")

class puppy(dog):
    def  weep(self):
        print("dog weeps")

p1 = puppy()
p1.sound()
p1.bark()
p1.weep()
```

### 6) Multiple Inheritance

### Create:

class Father
class Mother
class Child(Father, Mother)

### Display qualities inherited from both parents

```python
class father:
    def skill1(self):
        print("Father : coding")

class mother:
    def skill2(self):
        print("mother : Painting")

class child(father, mother):
    def skill3(self):
        print("child : dancing")

c1 = child()
c1.skill1()
c1.skill2()
c1.skill3()
```

### 7) Method Overriding

Create parent class Bird with method sound().

Create child class Crow that overrides the same method.

```python
class bird:
    def sound(self):
        print("bird makes sound")

class crow(bird):
    def sound(self):
        print("crow makes sound")

c1 = crow()
c1.sound()
```

### 8) Create a class Calculator with methods:

add()
subtract()
multiply()
divide()

Create an object and perform all four operations.

```python
class calculator:
    def add(self, a, b):
        return a+b
    
    def subtract(self, a, b):
        return a-b
    
    def multiply (self, a, b):
        return a*b
    
    def divide(self, a, b):
        if b != 0:
            return a/b
        else:
            return "division by zero not possible"
        
c1 = calculator()
print("Addition : ", c1.add(10,5))
print("subtraction: ", c1.subtract(30, 15))
print("multiplication: ", c1.multiply(10, 20))
print("division : ", c1.divide(50, 10))
    
```
