# Create a NumPy array and find sum & average

```python


import numpy as np

arr = ([10, 20, 30, 40, 50])

total = 0
for i in arr:
    total += i

avg = total / len(arr)

print("Sum: ", total)
print("Average: ", avg)


```

# Reverse a NumPy array

```python


import numpy as np 

arr = np.array([1, 2, 3, 4, 5])

rev = arr[:: -1]

print("Reversed Array: ", rev)

```

# Find maximum and minimum in NumPy array (no built-in max/min)

```python


import numpy as np

arr = np.array([45, 12, 89, 23, 67])

max_val = arr[0]
min_val = arr[0]

for i in arr: 
    if i > max_val:
        max_val = i
    if i < min_val:
        min_val = i

print("Maximum : ", max_val)
print("Minimum : ", min_val)
```
