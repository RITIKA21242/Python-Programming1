# Write a Python function to find the maximum and minimum numbers from a sequence of numbers.  (Note: Do not use built-in functions.)

```python


def find_max_min(arr):
    max_val = arr[0]
    min_val = arr[0]

    for i in arr:
        if i > max_val:
            max_val = i
        if i < min_val:
            min_val = i

    return max_val, min_val

numbers = [10, 20, 30, 40, 50]
print(find_max_min(numbers))
```

# Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.  

```python


def sum_of_cubes(n):
    sum = 0
    for i in range(1, n):
        sum += i ** 3
    return sum

print(sum_of_cubes(5))
```

# Write a Python function to print 1 to n using recursion.


```python

def print_numbers(n):
    if n == 0:
        return
    
    print_numbers(n-1)
    print(n)

print_numbers(5)
```

# Write a recursive function to print Fibonacci series upto n terms.

```python


def fibonacci(n):
    if n <= 1:
        return n 
    return fibonacci(n-1) + fibonacci(n-2)

n = 7
for i in range(n):
    print(fibonacci(i), end="")
```
