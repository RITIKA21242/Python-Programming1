# Write a function to display characters at even index positions.

```python


def even_index_chars(s):
    for i in range(0, len(s), 2):
        print(s[i], end = "")

even_index_chars("python")
```

# Write a program to extract all digits from a string.


```python

def extract_digits(s):
    digits = ""
    for ch in s:
        if ch.isdigit():
            digits += ch
    return digits

print(extract_digits("abcdef12345jbibj2079"))
```

# Find union, intersection, and difference of two sets

```python


def set_operations(a, b):
    return a | b, a & b, a - b

s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}
print(set_operations(s1, s2))
```

# Recursive function to find sum of first n natural numbers


```python

def sum_n(n):
    if n == 0:
        return 0
    return n + sum_n(n-1)

print(sum_n(5))
```
