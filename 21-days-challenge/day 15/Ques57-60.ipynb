# Lambda function to find volume of a cone


```python

volume_cone = lambda r, h: (1/3) * 3.14159 * r * r * h
print(volume_cone(3, 5))
```

# Lambda function to return (max, min) from a list

# Sample input: [10, 6, 8, 90, 12, 56]
# Sample output: (90, 6)

```python


max_min = lambda lst: (
    (lambda mx, mn: (mx, mn))(
        (lambda: [x for x in lst if all(x >= y for y in lst)][0])(),
        (lambda: [x for x in lst if all(x <= y for y in lst)][0])()
    )
)

print(max_min([10, 6, 8, 90, 12, 56]))
```

# Default Argument

# A value assigned beforehand, used if no argument is passed.

```python


def student(name, age):
    print(name, age)

student(age=18, name="Ritika")

def greet(name = "Guest"):
    print("Hello", name)

greet()
greet("Ritika")
```

# Variable Length Argument

# Used when number of arguments is unknown.

```python


def total(*numbers):
    s = 0
    for n in numbers:
        s += n
    return s

print(total(1, 2, 3, 4))
```
