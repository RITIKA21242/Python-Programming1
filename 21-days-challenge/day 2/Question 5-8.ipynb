Program to input a string and count length of the string

```python
s= input("enter a string:")
count=0
for ch in s:
    count+=1
print("length of string is:",count)
```

Program to input a string and check vowels in the string

```python
s= input("enter a string:")
vowel=0
for ch in s: 
    if ch in 'aeiouAEIOU':
        vowel+=1
print("number of vowels in string is:",vowel)
```

Program to input a string and reverse it

```python
s=input("enter a string: ")
rev= ""
for ch in s: 
    rev = ch + rev
print("reversed string is:",rev)
```
