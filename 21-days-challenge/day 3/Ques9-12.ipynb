# Calculating mathematical operations on x and y using arithematic operators

```python
# Calculating mathematical operations on x and y using arithematic operators
x=9
y=7
print("Addition of x and y is: ", x + y)
print("Subtraction of x and y is: ", x - y)
print("Multiplication of x and y is: ", x * y)
print("Division of x and y is: ", x / y)
```

# calculate area of circle using user input

```python
# calculate area of circle using user input
r=float(input("enter radius: "))
area = 3.14 * r * r
print("Area of circle is : ", area)
```

# calculate simple interest using user input for principle, rate and time

```python
# calculate simple interest using user input for principle, rate and time
p = float(input("enter principle: "))
r = float(input("enter rate: "))
t = float(input("enter time: "))
SI = (p * r * t) / 100
print ("Simple interest is : ", SI)
```

# calculate hypotenuse using user input for base and height

```python
# calculate hypotenuse using user input for base and height
a= float (input("enter height: "))
b= float (input("enter base: "))
c= ( a * a + b * b) ** 0.5
print("hypotenuse is: ", c)
```
