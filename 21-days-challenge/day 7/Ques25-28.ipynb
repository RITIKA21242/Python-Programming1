# Program to check whether a number is Palindrome


```python

n = int(input("enter a number: "))

temp = n
rev = 0
while temp > 0: 
    digit = temp % 10
    rev = rev * 10 + digit
    temp = temp //10

if rev ==n:
    print("Palindrome number")
else: 
    print("Not a Palindrome number")
```

# Program to find sum of digits of a number

```python


n = int(input("enter a number: "))
sum = 0
while n > 0 :
    sum = sum +n % 10
    n = n // 10

print("Sum of digits: ", sum)
```

# Count and print numbers divisible by 5 or 7 between 1 to 100

```python


count = 0
for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end = "")
        count += 1

print("\n Count :", count)
```
