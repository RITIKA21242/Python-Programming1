# Count occurrences of a substring (Left to Right)


```python

s= input("enter a string : ")
sub = input("enter a substring: ")

count = 0
for i in range(len(s) - len(sub) + 1):
    if s[i:i+len(sub)] == sub: 
        count +=1

print(count)
```

# Count occurrences of each alphabet (case-insensitive)


```python

s= input("enter a string: ").upper()
d={}

for ch in s:
    if ch.isalpha():
        d[ch] = d.get(ch, 0) + 1

for k in sorted(d):
    print(d[k], k, sep="")
```

# Count occurrences of each alphabet (case-insensitive)


```python

sentence = input("enter sentence: ")

words = sentence.split()

unique_words= set(words)
print("number of unique words: ", len(unique_words))
```

# Set operations on fruits

```python


s1 = set(input("enter fruits for set 1: ").split())
s2 = set(input("enter fruits for set 2: ").split())

print("fruits in both s1 and s2:", s1 & s2)
print("fruits only in s1:", s1-s2)
print("count of all fruits:", len(s1 | s2))
```
