# Scan n values (0–3) and count occurrences

```python


n = int(input("enter number of values: "))
count = [0, 0, 0, 0]

for _ in range(n):
    x=int(input())
    if 0 <= x <= 3: 
        count[x] += 1

for i in range(4):
    print(f"{i} occured {count[i]} times")
```

# Find runner-up score

```python


n = int(input("enter number of students: "))
scores = list(map(int, input("enter scores: ").split()))

unique_scores = list(set(scores))
unique_scores.sort(reverse=True)

if len(unique_scores) > 1:
    print(unique_scores[1])
else:
    print("No runner-up score")
```

# Dictionary of students and cities

```python


n = int(input("enter number of students: "))
d={}

for _ in range(n):
    name = input("enter name: ")
    city = input("enter city: ")
    d[name] = city

print("Names : ")
for name in d:
    print(name)

print("Cities: ")
for city in d.values(): 
    print(city)


print("Student Details : ")
for name, city in d.items():
    print(name, "-", city)

city_count = {}
for city in d.values():
    city_count[city] = city_count.get(city, 0) + 1

print ("Students per city: ")
for city, count in city_count.items():
    print(city, ":", count)
```
