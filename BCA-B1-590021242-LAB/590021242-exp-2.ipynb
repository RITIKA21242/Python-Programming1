8. Area of triangle when sides are given (Heronâ€™s formula)

```python
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

s = (a + b + c) / 2
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5

print("Area of triangle:", area)
```

9. Convert seconds into hours, minutes and seconds

```python
seconds = int(input("Enter total seconds: "))

hours = seconds // 3600
minutes = (seconds % 3600) // 60
remaining_seconds = seconds % 60

print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", remaining_seconds)
```

10. Swap two numbers without using third variable

```python
a = int(input("Enter a: "))
b = int(input("Enter b: "))

a, b = b, a

print("After swapping:")
print("a =", a)
print("b =", b)
```

12. Truth table for bitwise operators (&, |, ^)

```python
a = 1
b = 0

print("a b a&b a|b a^b")
print(a, b, a & b, a | b, a ^ b)
print(a, a, a & a, a | a, a ^ a)
print(b, b, b & b, b | b, b ^ b)
print(b, a, b & a, b | a, b ^ a)
```
