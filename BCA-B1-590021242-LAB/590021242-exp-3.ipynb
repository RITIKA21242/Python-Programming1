### 1. Check whether a given number is divisible by 3 and 5 both. 

```python
n = int(input("Enter a number: "))

if n % 3 == 0 and n % 5 == 0:
    print("Divisible by both 3 and 5")
else:
    print("Not divisible by both 3 and 5")
```

### 3. Find the greatest among two numbers. If numbers are equal than print “numbers are equal”.  

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("Greatest number:", a)
elif b > a:
    print("Greatest number:", b)
else:
    print("Numbers are equal")
```

### 4. Find the greatest among three numbers assuming no two values are same.  

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a > b and a > c:
    print("Greatest number:", a)
elif b > a and b > c:
    print("Greatest number:", b)
else:
    print("Greatest number:", c)
```

### 5. Check whether the quadratic equation has real roots or imaginary roots. Display the roots. 

```python
import math

a = int(input("Enter a: "))
b = int(input("Enter b: "))
c = int(input("Enter c: "))

d = b*b - 4*a*c

if d > 0:
    r1 = (-b + math.sqrt(d)) / (2*a)
    r2 = (-b - math.sqrt(d)) / (2*a)
    print("Real and distinct roots:", r1, r2)
elif d == 0:
    r = -b / (2*a)
    print("Real and equal roots:", r)
else:
    print("Imaginary roots")
```

### 6. Find whether a given year is a leap year or not. 

```python
year = int(input("Enter year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not a leap year")
```

### 7. Write a program which takes any date as input and display next date of the 
### calendar e.g.  
### I/P: day=20 month=9 year=2005   
### O/P: day=21 month=9 year 2005 

```python
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

if day < 30:
    day += 1
else:
    day = 1
    if month < 12:
        month += 1
    else:
        month = 1
        year += 1

print("Next date:", day, month, year)
```

0 to 3.4 -> F  
3.5 to 5.0->C+  
5.1 to 6->B  
6.1 to 7-> B+  
7.1 to 8-> A  
8.1 to 9->A+  
9.1 to 10-> O (Outstanding)

Subject name: Marks  
PDS: 70

```python
name = input("Enter name: ")
roll = input("Enter roll number: ")

m1 = int(input("Enter marks 1: "))
m2 = int(input("Enter marks 2: "))
m3 = int(input("Enter marks 3: "))
m4 = int(input("Enter marks 4: "))
m5 = int(input("Enter marks 5: "))

total = m1 + m2 + m3 + m4 + m5
percentage = total / 5
cgpa = percentage / 10

if cgpa <= 3.4:
    grade = "F"
elif cgpa <= 5.0:
    grade = "C+"
elif cgpa <= 6.0:
    grade = "B"
elif cgpa <= 7.0:
    grade = "B+"
elif cgpa <= 8.0:
    grade = "A"
elif cgpa <= 9.0:
    grade = "A+"
else:
    grade = "O"

print("Percentage:", percentage)
print("CGPA:", cgpa)
print("Grade:", grade)
```
