### Q1.	Add few names, one name in each row, in “name.txt file”. 

- Count no of names
- Count all names starting with vowel 
- Find longest name 


### Code Explainations: Count no of names 

1. `file = open("name.txt", "r")` : opens txt file named "name.txt" in read only mode
2. `names = file.readlines()` : reads all the lined entered in the text file
3. `print("Total Numbers: ", len(names))` : returns length or count of lines in the txt file 
4. `file.close()` : closes the file 

```python
file = open("name.txt", "r")
names = file.readlines()
print("Total Numbers: ", len(names))
file.close()
```

### Code Explainations : count all names with vowel 

- `f = open("name.txt", "r")`: opens the name.txt file in read only mode 

- `count = 0` : initial value of count is 0 

- `for name in f:`: condition applied

- ` if name.strip()[0].lower() in 'aeiou':` : here `strip()` removes new line character, `[0]` returns first character and `lower()` handles capital letters

```python
f = open("name.txt", "r")
count = 0

for name in f:
    if name.strip()[0].lower() in 'aeiou':
        count +=1

print("Total: ", len(name))
f.close()
```

### Code Exaplainations : Print the longest name 

1. `longest = ""` : creates an empty string

2. `for name in f:  if len(name) > len(longest):` : sets the condition if the name is greater than the length of string which by default is empty for now 

3. `longest=name` : prints the extracted value(name) into empty string longest

```python
f = open("name.txt", "r")
longest = ""

for name in f:
    if len(name) > len(longest):
        longest = name

print("Longest name: ", longest)
f.close()
```

### Q2. Store integers in a file. 
- Find the max number 
- Find average of all numbers 
- Count number of numbers greater than 100 


### Code Explainations: Max number

1. `file = open("numbers.txt", "r")` : opens number.txt file in read mode

2. `numbers = [int(x) for x in file]` : `int(x)` converts string into integer and `for x in file` reads all the lines in the txt file

3. `print("Maximum number:", max(numbers))` : using max function prints the maximum no in the txt file


```python
file = open("numbers.txt", "r")
numbers = [int(x) for x in file]
print("Maximum number:", max(numbers))
file.close()
```

### Code Explainations : Average of numbers

1. `f = open("numbers.txt", "r")` : opens the file in read mode

2. `number = [int(x) for x in f]`: converts string to integer and reads all the lines

3. `avg = sum(number) / len(number)` : calculates the average by adding all the numbers divided by the no of no present in the txt file 


```python
f = open("numbers.txt", "r")
number = [int(x) for x in f]
avg = sum(number) / len(number)
print("Average : ", avg)
f.close()
```

### Code Exaplainations : Count of no greater than 100 

1. `count = 0` : sets the initial value of count be 0

2. `for num in f:` : reads all the numbers

3. `if int(num) > 100:` : converts num into integer and condition is set to be num greater than 100

4. `count += 1` : the count increments what number is stored in num 

```python
f= open("numbers.txt", "r")
count = 0

for num in f:
    if int(num) > 100:
        count += 1

print("Count  of number : ", num)
f.close()
```

### Q3. 3.	Assume a file city.txt with details of 5 cities in given format (cityname population(in lakhs) area(in sq KM) ): 

#### Example: 

Dehradun 5.78 308.20 

Delhi 190 1484 

#### Open file city.txt and read to:

- Display details of all cities 
- Display city names with population more than 10Lakhs 
- Display sum of areas of all cities 


### Code Explainations : Details of cities

- `open()` → opens file in read mode

- `for line` in file: → reads each city record

- `split()` → breaks line into 3 parts

- Values are stored in variables

- Printed in readable format

```python
f = open("city.txt", "r")

for line in f:
    city, pop, area = line.split()
    print("City:", city, "Population:", pop, "Area:", area)

f.close()
```

### Code Explainations: city names and their area more than 10 lakhs 

- `open()` : opens file in read mode

- `for line in f:` : reads all line

- `if float(pop) > 10:` : pop is converted into float and area greater than 10 lakh condition is made

```python
f = open("city.txt", "r")
for line in f:
    city, pop, area = line.split()
    if float(pop) > 10:
        print(city)
f.close()
```

### Code Explainations : sum of all cities area

1. `total_area = 0` : the initial value of total area is said to be 0

2. `for line in f:` : all lines inside the txt file is will get read

3. `total_area += float(area)` : adds all the cities area and stores them in total_area

```python
f = open("city.txt", "r")
total_area = 0

for line in f:
    city, pop, area = line.split()
    total_area += float(area)

print("Total Area : ", total_area)

f.close()
```

### Q4. 4.	Input two values from user where the first line contains N, the number of test cases. The next N lines contain the space separated values of a and b. Perform integer division and print a/b. Handle exception in case of ZeroDivisionError or ValueError.  

Sample input 

1 0 

2 $ 

3 1  

Sample Output: 

Error Code: integer division or modulo by zero 
 
Error Code: invalid literal for int() with base 10: '$' 3 


### Code Explainations :

1. `n = int(input())` : takes the number of test cases from the user

2. `for i in range(n)`: : runs the loop for the given number of test cases

3. `a, b = input().split()` : reads two space-separated values

4. `a = int(a)` : converts the first value into integer

5. `b = int(b)` : converts the second value into integer

6. `print(a // b)` : performs integer division and displays the result

7. `try:` : encloses the code that may generate runtime errors

8. `except ZeroDivisionError`: : handles the case when divisor is zero

9. `except ValueError:` : handles invalid input that cannot be converted to integer

```python
try:
    n = int(input())

    for i in range(n):
        a, b = input().split()
        a = int(a)
        b = int(b)
        print(a // b)

except ZeroDivisionError:
    print("Error Code: integer division or modulo by zero")

except ValueError as e:
    print("Error Code:", e)
```

### Q5.Create multiple suitable exceptions for a file handling program.

### Code Explainations : 

- `try:` : encloses the file handling code that may raise exceptions

- `file = open("data.txt", "r")` : opens the file data.txt in read mode

- `print(file.read())` : reads and displays the entire content of the file

- `except FileNotFoundError:` : handles the error when the specified file is not found

- `print("Error: File not found") `: displays an appropriate error message

- `except PermissionError:` : handles the error when file access permission is denied

- `print("Error: Permission denied") `: displays permission-related error message

- `finally:` : block that executes in all cases

- `print("Program execution completed")` : indicates successful termination of the program

```python
try:
    file = open("data.txt", "r")
    print(file.read())

except FileNotFoundError:
    print("Error: File not found")

except PermissionError:
    print("Error: Permission denied")

finally:
    print("Program execution completed")
```
