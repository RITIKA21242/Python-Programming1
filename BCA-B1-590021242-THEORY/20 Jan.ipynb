### Varriables - Native Python

A variable in Python is a name that refers to a value stored in memory. 
The value assigned to a variable can be changed during the execution of a program.
Variables allow programs to store, update, and manipulate data efficiently.


### Naming and Using Variables

#### Guidelines

- Variable names may contain letters, digits, and underscores. They must begin with a letter or an underscore and cannot start with a number.
- Spaces are not permitted in variable names. To separate words, underscores should be used instead.
- Reserved keywords and built-in function names should not be used as variable names, as they have predefined meanings in Python.
- Variable names should be concise yet meaningful to clearly represent the stored data.
- Avoid using characters that look similar to numbers, such as lowercase `l` and uppercase `O`, to prevent confusion.

---

> **Note:**  
> It is recommended to use lowercase letters while naming variables.  
> Although uppercase letters do not cause errors, using lowercase improves readability and consistency.

### 1) <span style="color:#E6C7F0; font-weight:bold;">Case Conversion Methods
</span> 

`lower()`

#### Definition and Usage

The lower() method returns a new string in which all uppercase characters are converted to lowercase.

---

```python
"PYTHON".lower()
```

---

`upper()`

#### Definition and Usage

The upper() method returns a new string where all lowercase characters are converted to uppercase

```python
"python".upper()
```

---

`title()`

#### Definition and Usage

The title() method converts the first letter of every word in a string to uppercase and the remaining letters to lowercase.

```python
text = "python programming language"
print(text.title())
```

---

`capitalize()`

#### Definition and Usage

The capitalize() method converts only the first character of the string to uppercase and converts all other characters to lowercase.

```python
sentence = "pYTHON is FUN"
print(sentence.capitalize())
```

---

`swapcase()`

#### Definition and Usage

The swapcase() method changes uppercase letters to lowercase and lowercase letters to uppercase in a string.

```python
word = "PyThOn"
print(word.swapcase())
```

---

`casefold()`

#### Definition and Usage

The casefold() method converts a string into a lowercase format suitable for case-insensitive comparisons. It is stronger than lower().

```python
text = "GERMAN ÃŸ"
print(text.casefold())
```

### <span style="color:#E6C7F0; font-weight:bold;"> 2. Remove Extra Spaces
</span>

`strip()`

The strip() method removes all leading and trailing whitespace characters from a string. It is commonly used to clean user input and formatted text.

```python
text = "   Python Programming   "
print(text.strip())
```

---

`lstrip()`

The lstrip() method removes whitespace characters only from the left side (beginning) of a string, leaving the right side unchanged.

```python
text = "   Python Programming   "
print(text.lstrip())
```

---

`rstrip()`

The rstrip() method removes whitespace characters only from the right side (end) of a string, leaving the left side unchanged.

```python
text = "   Python Programming   "
print(text.rstrip())
```
