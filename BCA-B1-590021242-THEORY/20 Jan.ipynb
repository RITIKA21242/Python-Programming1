# Introduction to Python

## Python

Python is a high-level, general-purpose programming language known for its simple syntax and readability. It allows programmers to focus more on problem-solving rather than complex syntax rules. Python supports multiple programming paradigms such as procedural, object-oriented, and functional programming, making it suitable for a wide range of applications including web development, data analysis, and automation.

##  what is Interactive Programming in Python?

Interactive programming refers to executing Python statements directly and receiving immediate output. In this mode, instructions are executed line by line, making it highly useful for testing small pieces of code

## What is script programming in python?

Script programming involves writing Python code in a file and executing the entire program at once. The script is saved with a .py extension and executed using a command or environment.

### What are immutable data types()?

Data types in Python are classified based on whether their values can be modified after creation.

Mutable Data Types: These allow changes to their content (e.g., lists, dictionaries).

Immutable Data Types: These do not allow modification once created (e.g., integers, strings, tuples).
