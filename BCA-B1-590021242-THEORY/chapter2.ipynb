# Lecture 1 — Strings
## 1) Initialization
- Strings are immutable sequences of characters.

- Create using:

  - Single quotes: 'hello'
  - Double quotes: "hello"
  - Triple quotes (multi-line): """hello"""

```python
s1 = 'Python'
s2 = "Data Science"
s3 = """Line1
Line2"""

print(s1)
print(s2)
print(s3)
```

### Operators on strings

- `+` concatenation, `*` repetition
- `in` membership
- comparison uses lexicographic order


```python
print("Py" + "thon")
print("ha" * 3)
print("a" in "cat")
print("Abc" < "abd")
```

```python
l = ["Ritika", "Singh"]
"Ritika" in l
```

### 3) Indexing (0-based)

```python
s = "Python"
print(s[0])
print(s[-1])
print(s[len(s)-1])
print(s[len(s)])  # s[len(s)] = s[6] and in the index 6 does not exist as len(n) by default starts with 0 nd end is len(n)-1
```

### 4) Slicing 

Form: `s[start : stop : step] (stop excluded)`

```python
s = "Python"
print(s[1:4])
print(s[:3])
print(s[3:])
print(s[::-1])
print(s[5:1:-1])
```

### 5) `split()`

Splits into list by delimmiter (default = whiitespace)

```python
print("Python is a programming languagr".split())
print("1 2 3 4 5 6 ".split())
```

### 4) List Comprehension

compact way to build lists.

```python
for item in iterator:
    return item*item
```

```python
squares = [x*x for x in range(1,6)]
evens = [x for x in range(10) if x%2==0]
pairs = [(x,y) for x in [1,2] for y in [3,4,6]]
print(squares)
print(evens)
print(pairs)
```

## Advanced Functions

### 1) `lambda`
Anonymous function

```python
def square(x):
    return x*x
```

```python
square(2)
```

```python
l = [1, 2, 3, 4]

def square(x):
    return x*x
new_list=[]
for each in l: 
    new_list.append(square(each))
```

```python
[x*x for x in l]
list(map(lambda x : x*x, l))
```

```python
def square(number):
    return number * number

numbers = [1, 2, 3, 4, 5]

squared_numbers = map(square, numbers)

print(list(squared_numbers))
```

```python
list(map(square, l))
```

### 2) `map()`

Apply a function to every element 

```python
nums = [1,2,3]
list(map(lambda x: 100/x, nums))
```

### 3) `filter()`

Keep only elements satisfying conditions

```python
nums = [1,2,3,4,5]
list(filter(lambda x: x%2==0, nums)) 
```

### 4) Inner functions + returning functions

```python
def outer(msg):
    def inner():
        return msg.upper()
    return inner

f = outer("hello")
print(f)
print(f()) 
```

```python
outer("help")()
```

### Passing collections/functions


```python
def apply_twice(f, x):
    return f(f(x))

apply_twice(lambda a: a+1, 5)
```

## Mutability in functions + Applications + Doubts
### 1) Mutable vs Immutable behavior
Immutable types: `int, float, str, tuple` Mutable types: `list, dict, set`

Key idea:
- If you modify a mutable object inside a function, the caller sees it.
- If you “modify” immutable objects, you actually create a new object.

```python
def change_num(x):
    x = x + 1

def change_list(L):
    L.append(99)

a = 10
change_num(a)
print(a)    

L = [1,2]
change_list(L)
print(L)  
```
