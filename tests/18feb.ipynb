### Stable Top-K Selection
The function should return the top k elements based on score.

### Input
items: List of tuples (name, score)
k: Integer
### Example:

items = [("A", 90), ("B", 95), ("C", 95), ("D", 85)]
k = 2
Expected Output
[("B", 95), ("C", 95)]
(Tie must preserve original order.)

### Given Code (Incorrect)

```python
def top_k(items, k):
    items.sort(key=lambda x: x[1], reverse=True)
    return items[:k]

items = [("A", 90), ("B", 95), ("C", 95), ("D", 85)]
k = 2
print(top_k(items, k))
```

### Question -2 
Category Word Counter
Count occurrences of words under each category.

### Input
pairs = [("sports","win"), ("sports","win"), ("tech","ai")]
Expected Output
{
    "sports": {"win": 2},
    "tech": {"ai": 1}
}

```python
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        d.setdefault(cat, {})
        d[cat].setdefault(word, 0)
        d[cat][word] += 1
    return d

pairs = [("sports","win"), ("sports","win"), ("tech","ai")]
print(count_by_category(pairs))
```

Missing Number (0 to n)
You are given n distinct numbers from range 0 to n. Return the missing number.

Input
nums = [3, 0, 1]
Expected Output
2

```python
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) // 2
    return int(total - sum(nums))

nums = [3, 0, 1]
print(missing_number(nums))
```

### Question -4
Default Argument Bug
The function should:

Return a list of unique values added.
If no second argument provided, each call should start fresh.
Input
add_unique(1)
add_unique(2)
Expected Output
Each call should return:

[1]
[2]

Given Code (Incorrect)

```python
def add_unique(x, seen=set()):
    if x not in seen:
        seen.add(x)
    return list(seen)

print(add_unique(1))
print(add_unique(2))

```

Question 5 (10 Marks)
Moving Average of Last w Elements
Write a function:

def moving_average(stream, w):
Input
stream: List of integers
w: Window size
Example:

stream = [1,2,3,4,5]
w = 3
Expected Output
Return list of averages:

[1.0, 1.5, 2.0, 3.0, 4.0]
(Each index shows average of last w elements up to that position.)

Must run in O(n).

```python
def moving_average(stream, w):
    result = []
    window_sum = 0

    for i in range(len(stream)):
        window_sum += stream[i]

        if i >= w:
            window_sum -= stream[i - w]

        current_window_size = min(i + 1, w)
        result.append(window_sum / current_window_size)

    return result

stream = [1,2,3,4,5]
w=3
print(moving_average(stream, w))
```
